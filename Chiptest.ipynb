```python
import pandas as pd
```

```python
df = pd.read_excel('chip_dataset_fix.xlsx')
df
```

```python
df.columns = ['No','Product','Type','ReleaseDate','RelaseDateFix','ProcessSize','TDP','DieSize','Transsistors','Freq','Foundry','Vendor','FP16','Fp32','FP64']
df
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
def_ReleaseDate = "0"
df.ReleaseDate.fillna(def_ReleaseDate, inplace =True)
def_ProcessSize = "0"
df.ProcessSize.fillna(def_ProcessSize, inplace =True)
def_TDP = "0"
df.TDP.fillna(def_TDP, inplace =True)
def_DieSize = "0"
df.DieSize.fillna(def_DieSize, inplace =True)
def_Transsistors = "0"
df.Transsistors.fillna(def_Transsistors, inplace =True)
def_FP16 = "0"
df.FP16.fillna(def_FP16, inplace =True)
def_Fp32 = "0"
df.Fp32.fillna(def_Fp32, inplace =True)
def_FP64 = "0"
df.FP64.fillna(def_FP64, inplace =True)
df
```

```python
df.isna().sum()
```

```python
df
```

```python
for index, row in df.iterrows():
    df.loc[index, 'TahunRelease'] = row['RelaseDateFix'][-4:]
df
```

```python
from datetime import datetime
now = datetime.now()

now.year
for index, row in df.iterrows():
    df.loc[index, 'LamaRelease_Tahun']=now.year - int (row['TahunRelease'])
df
```

```python
d_train = df[['No','Product','Type','TahunRelease','LamaRelease_Tahun']].reset_index(drop=True)
d_train
```

```python
for index, row in d_train.iterrows():
    if row['LamaRelease_Tahun'] >=7:
        d_train.loc[index, 'Keterangan_Release'] = 'lama'
    elif row['LamaRelease_Tahun'] >=6:
        d_train.loc[index, 'Keterangan_Release'] = 'sedang'
    else:
        d_train.loc[index, 'Keterangan_Release'] = 'Baru'
d_train
```
